```python
## Importing the required libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
## Reading the data 
df = pd.read_csv(r"zomato.csv")
```

```python
## Size of the data
df.shape
```

```python
## Top 5 rows of the data 
df.head()
```

```python
# Checking the null values
df.isnull().sum()
```

```python
# Replacing Delivery person age nan values with median value
value = df['Delivery_person_Age'].median()
df['Delivery_person_Age'] = df['Delivery_person_Age'].replace(np.nan,value)
```

```python
# Replacing Delivery_person_Ratings nan values with mean values
value = df['Delivery_person_Ratings'].mean()
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].replace(np.nan,value)
```

```python
# Replacing Weather condition nan values with most frequent values
value = df['Weather_conditions'].value_counts().index[0]
df['Weather_conditions'] = df['Weather_conditions'].replace(np.nan,value)
```

```python
# Replacing Road traffic density nan values with most frequent values
value = df['Road_traffic_density'].value_counts().index[0]
df['Road_traffic_density'] = df['Road_traffic_density'].replace(np.nan,value)
```

```python
# Replacing multiple deliveries nan values with most frequent values
value = df['multiple_deliveries'].value_counts().index[0]
df['multiple_deliveries'] = df['multiple_deliveries'].replace(np.nan,value)
```

```python
# Replacing Festival nan values with most frequent values
value = df['Festival'].value_counts().index[0]
df['Festival'] = df['Festival'].replace(np.nan,value)
```

```python
# Replacing City nan values with most frequent values
value = df['City'].value_counts().index[0]
df['City'] = df['City'].replace(np.nan,value)
```

```python
# Replacing Time_orderd_hour nan values with next valid observation to fill gap
df['Time_Orderd_hour'] = df['Time_Orderd_hour'].fillna(method='bfill')
```

```python
# Replacing Time_Orderd_min nan values with next valid observation to fill gap
df['Time_Orderd_min'] = df['Time_Orderd_min'].fillna(method='bfill')                 
```

```python
# Replacing Time_Order_picked_hour nan values with next valid observation to fill gap
df['Time_Order_picked_hour'] = df['Time_Order_picked_hour'].fillna(method='bfill')
```

```python
# Replacing Time_Order_picked_min nan values with next valid observation to fill gap
df['Time_Order_picked_min'] = df['Time_Order_picked_min'].fillna(method='bfill')
```

```python
# Checking the null values
df.isnull().sum()
```

```python
# Top 5 rows of the data 
df.head()
```

```python
# Applying one hot encoding on Type_of_order, Type_of_vehicle, City and days_of_week column
df = pd.get_dummies(df,columns=['Type_of_order','Type_of_vehicle','City','days_of_week'],drop_first=True)
```

```python
# Mapping Category of festival column 
df['Festival'] = df['Festival'].map({'No':'0','Yes':'1'})
```

```python
# Applying Ordinal encoding on Road_traffic_density column
df['Road_traffic_density'] = df['Road_traffic_density'].map({'Low':'0','Medium':'1','High':'2','Jam':'3'})
```

```python
# Applying Ordinal encoding on Weather_conditions column
df['Weather_conditions'] = df['Weather_conditions'].map({'Sunny':'0','Cloudy':'1','Fog':'2','Windy':'3','Sandstorms':'4','Stormy':'5'})
```

```python
# Applying one hot encoding on City_name column
df = pd.get_dummies(df,columns=['City_name'],drop_first=True)
```

```python
# Top 5 rows of the data
df.head()
```

```python
# Seperating Independent and Dependent feature
X = df.drop(labels=['Time_taken (min)'],axis=1)
y = df[['Time_taken (min)']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=32)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
## Train multiple models

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)

    #Make Predictions
    y_pred=model.predict(X_test)

    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
